```python
import keras
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import os
```

# Settings

```python
size = 256
seed = 5
```

```python
test_generator = ImageDataGenerator(rescale = 1./255)

test_data_generator = test_generator.flow_from_directory(
                    'drive/MyDrive/project_data_V1/test', 
                     target_size=(size, size),
                     batch_size=32,
                     shuffle=False,
                     seed = seed)
```

# Model

```python
model = keras.models.load_model('drive/MyDrive/models/cnn_bestmodel.h5')
```

# Predicting probabilities

```python
probabilities = model.predict(test_data_generator)
```

# Plotting results

```python
encoding = test_data_generator.class_indices
decoding = dct = {v: k for k, v in encoding.items()}
image_genre = [decoding[label] for label in test_data_generator.labels]
```

```python
top3_genre_proba_list = []
for prob in probabilities:
  top3_genre_proba_dict = {}
  for _ in range(3):
    max_i = np.argmax(prob)
    top3_genre_proba_dict[decoding[max_i]] = prob[max_i]
    prob[max_i] = 0.0
  top3_genre_proba_list.append(top3_genre_proba_dict)
```
